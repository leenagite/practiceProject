```python
import pandas as pd 
import numpy as np

import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
data['Country'] = le.fit_transform(data['Country'])
data['Country']
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
data['Region'] = le.fit_transform(data['Region'])
data['Region']
```

```python
data.info()
```

```python
data
```

```python
y = data.iloc[:,3]
```

```python
y
```

```python
x = data.iloc[:,:]
del data['Happiness Score']
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=12345)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
x_train.head()
```

```python
from sklearn import linear_model
```

```python
regr = linear_model.LinearRegression()
```

```python
regr.fit(x_train, y_train)
```

```python
print ('coeffienct',regr.coef_)
print ('Intercept',regr.intercept_)
```

```python
expected = y_test
```

```python
y_pred = regr.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
print(r2_score(expected, y_pred))
mean_squared_error(expected, y_pred)
```

```python

from sklearn.model_selection import cross_val_score
accuracies = cross_val_score(estimator= regr, X=x, y=y, cv=20)
accuracies

#print("Accuracy Mean {} Accuracy variance \ {}".format(accuracies_mean(), accuracies_std()))
```
